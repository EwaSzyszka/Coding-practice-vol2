```python
from scipy.stats import gamma
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
a = 1.99
a2 = 3
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')

x = np.linspace(gamma.ppf(0.01, a),gamma.ppf(0.99, a), 100)
ax.plot(x, gamma.pdf(x, a), 'b-', lw=1, alpha=0.5, label='gamma a')
ax.plot(x, gamma.pdf(x, a2), 'r-', lw=1, alpha=0.5, label='gamma pdf')


r = gamma.rvs(a, size=100)
r2 = gamma.rvs(a2, size=100)

ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
ax.hist(r2, normed=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
from scipy.stats import beta
fig, ax = plt.subplots(1, 1)
a, b = 2, 0.6
a2,b2 = 6,7
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
ax.plot(x, beta.pdf(x, a, b), 'b-', lw=1, alpha=0.7, label=' a = 2, b = 0.6')
ax.plot(x, beta.pdf(x, a2, b2), 'r-', lw=1, alpha=0.7, label='a = 6, b = 7')

ax.legend(loc='best', frameon=False)

#random variables
r = beta.rvs(a, b, size=100)
r2 = beta.rvs(a2, b2, size=100)

ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
ax.hist(r2, normed=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```
