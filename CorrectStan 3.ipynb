```python
stan_code = """

data {

    int<lower=1> N; //number of prices recorded: len(price)
    int<lower=1> T; //number of types of products: 10
    int<lower=1> S; //number of stores: 4
    
    real alpha; //needed hyperparameter for lambda
    real beta;//needed hyperparameter for lambda
    
    real<lower=0> prices[N]; //observed price of each product 
    int<lower=0> store_type[N]; //type of store
    int<lower=0> product_type[N]; //product type 
    int<lower=0> brand[N]; //branded vs not-branded product 
    
}

// things we cannot see are the parameters

parameters {
    real <lower=0> base_price[T];  //The model will estimate the base price of each product 
    real <lower=0> multiplier_store[S]; //The model will estimate the multiplier of each store
    real <lower=0> lambda;  //unknown, sampled from gamma distribution with fixed alpha and beta
    real <lower =0> sigma2; //used for the normal distribution that generates prices
    
}


//likelihood function
//justifying the priors 

model {
    lambda ~ gamma(alpha,beta);  //gamma prior 
    
    for(i in 1:T) {
    base_price[product_type[i]] ~ exponential(lambda); //exponential prior, as it is positive 
    }
    for(i in 1:S) {
    multiplier_store[store_type[i]] ~ exponential(lambda);  
    }
    //all the prices that I have
    for(i in 1:N) { 
        prices[i] ~ normal(base_price[product_type[i]]* multiplier_store[store_type[i]], sigma2) ; 
    }
    
    }
"""

stan_model = pystan.StanModel(model_code=stan_code)
```

```python
stan_code = """

data {

    int<lower=1> N; //number of prices recorded 
    int<lower=1> T; //number of types of products 10
    int<lower=1> S; //number of stores 1,2,3,4
    int<lower=1> L; //number of locations: 1,2,3,4...
    int<lower=1> B; //number of brands: branded versus not branded
    
    real alpha; //needed for the hyperparameters for lambda
    real beta;//needed for the hyperparameters for lambda
    
    real<lower=0> prices[N]; //observed price of each product 
    int<lower=0> store_type[N]; //type of store
    int<lower=0> product_type[N]; //product type 
    
}

// things we cannot see are the parameters

parameters {
    real <lower=0> base_price[T]; //10   
    real <lower=0> multiplier_store[S]; //4
    real <lower=0> location[L]; //
    real <lower=0> brand[B]; //
    real <lower=0> lambda;  //hyperparameters that are not fixed and I generate
    real <lower =0> sigma2;
    
}


model {
    lambda ~ gamma(alpha,beta);  //gamma prior 
    
        
    for(i in 1:T) {
    base_price[i] ~ exponential(lambda); //exponential prior, as it is positive 
    
    }
    
    for(i in 1:L){
    location[i]~ exponential(lambda);
    }
    
    
    for(i in 1:S) {
    multiplier_store[store_type[i]] ~ exponential(lambda);  
    }
    
    for(i in 1:B){
    brand[i]~ exponential(lambda);
    }

    //all the prices that I have
    for(i in 1:N) { 
        prices[i] ~ normal(base_price[product_type[i]]* multiplier_store[store_type[i]], sigma2) ; 
    }
    
    }         
"""

stan_model = pystan.StanModel(model_code=stan_code)
```
